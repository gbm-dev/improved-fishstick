To add vectors we add each cooresponding component.
Example:

\begin{bmatrix} 0 \\ 2 \end{bmatrix} + \begin{bmatrix} 3 \\ 4 \end{bmatrix} = \begin{bmatrix} 0+3 \\ 2+4 \end{bmatrix} = \begin{bmatrix} 3 \\ 6 \end{bmatrix}




```python
# Define the given vectors
vec1 = np.array([0,2])
vec2 = np.array([3,4])

# Add the vectors
result_vec = vec1 + vec2

# Plot the vectors and their sum
plt.figure()
plt.quiver(0, 0, vec1[0], vec1[1], angles='xy', scale_units='xy', scale=1, color='r', label="Vector 1")
plt.quiver(vec1[0], vec1[1], vec2[0], vec2[1], angles='xy', scale_units='xy', scale=1, color='b', label="Vector 2")
plt.quiver(0, 0, result_vec[0], result_vec[1], angles='xy', scale_units='xy', scale=1, color='g', label="Resultant Vector")

# Set graph limits and labels
plt.xlim(-1, 8)
plt.ylim(-1, 8)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a legend and grid
plt.legend()
plt.grid()

# Show the graph
plt.show()
```

Again a function can be made to easily reuse the code.

```python
def plot_add_vectors(vec1, vec2):
    result_vec = vec1 + vec2
    
    # Plot the vectors and their sum
    plt.figure()
    plt.quiver(0, 0, vec1[0], vec1[1], angles='xy', scale_units='xy', scale=1, color='r', label="Vector 1")
    plt.quiver(vec1[0], vec1[1], vec2[0], vec2[1], angles='xy', scale_units='xy', scale=1, color='b', label="Vector 2")
    plt.quiver(0, 0, result_vec[0], result_vec[1], angles='xy', scale_units='xy', scale=1, color='g', label="Resultant Vector")

    # Set graph limits and labels
    plt.xlim(-1, max(vec1[0] + vec2[0], vec2[0]) + 1)
    plt.ylim(-1, max(vec1[1] + vec2[1], vec2[1]) + 1)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    # Add a legend and grid
    plt.legend()
    plt.grid()

    # Show the graph
    plt.show()
```

```python
vec1 = np.array([1 ,5])
vec2 = np.array([3 ,9])

plot_add_vectors(vec1, vec2)
```
