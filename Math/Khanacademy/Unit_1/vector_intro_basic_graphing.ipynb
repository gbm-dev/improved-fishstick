```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

It is best practice to include all imports at the beginning of the notebook in a cell dedicated to imports.

```python
vector = np.array([5, 0])
origin = np.array([0, 0])

plt.quiver(*origin, *vector, angles='xy', scale_units='xy', scale=1, color='r', label='(5,0)')
plt.xlim(-1, 6)
plt.ylim(-1, 6)

plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()

plt.show()
```

Lets describe some of the functions used above.

1. plt.quiver(): This function is used to plot a vector as an arrow on a 2D plot. It takes the x and y coordinates of the starting point of the arrow, and the x and y components of the vector as arguments, along with several other optional arguments to control the appearance of the arrow, such as its color, size, and style.

2. plt.xlim(): This function is used to set the x-axis limits of the plot. In this code, we set the limits to -1 and 6, which means the x-axis will range from -1 to 6.

3. plt.ylim(): This function is used to set the y-axis limits of the plot. In this code, we set the limits to -1 and 6, which means the y-axis will range from -1 to 6.

4. plt.grid(): This function is used to add a grid to the plot. When called with no arguments, it adds a default grid to the plot.

5. plt.xlabel(): This function is used to set the label for the x-axis of the plot. In this code, we set the label to 'X-axis'.

6. plt.ylabel(): This function is used to set the label for the y-axis of the plot. In this code, we set the label to 'Y-axis'.

7. plt.legend(): This function is used to add a legend to the plot. In this code, we included a label for the vector using the label argument of plt.quiver(), and the plt.legend() function displays the legend.

8. plt.show(): This function is used to display the plot. When called, it shows the plot in a separate window or cell output, depending on the configuration of your notebook environment.

The above graph is nice but I would like to include some labels on the vectors.

```python
vector = np.array([5, 0])
origin = np.array([0, 0])

plt.quiver(*origin, *vector, angles='xy', scale_units='xy', scale=1, color='y', label='(5,0)') # Colors can be changed by modifying the color='y'
plt.text(2.5, 0.5, "v", fontsize=14, color='y', ha='center', va='center')

plt.xlim(-1, 6)
plt.ylim(-1, 6)

plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()

plt.show()
```

This includes a label but what happens if the vector changes?

```python
vector = np.array([5, 3])
origin = np.array([0, 0])

# Define an RGB color as a tuple of three values between 0 and 1.
color = (0.2, 0.4, 0.6)

plt.quiver(*origin, *vector, angles='xy', scale_units='xy', scale=1, color=color, label='(5,0)')
plt.text(2.5, 0.5, "v", fontsize=14, color=color, ha='center', va='center')

plt.xlim(-1, 6)
plt.ylim(-1, 6)

plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()

plt.show()
```

This code is getting very verbose. We have to type the entire snippet everytime we want to make a change to the graph. Let's wrap the code in a function.

```python
def plot_vector(vector, origin, label=None):
    # Modifying the color to randomize each time the function is called.
    r = random.random()
    g = random.random()
    b = random.random()
    color = (r, g, b)
    plt.quiver(*origin, *vector, angles='xy', scale_units='xy', scale=1, color=color, label=label)

 


    if label:
        plt.text(*vector + 0.2, label, fontsize=14, color=color) # This will put the label at the endpoint of the vector instead of the label being statically set.

    plt.xlim(-1, 6)
    plt.ylim(-1, 6)

    plt.grid()
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    plt.legend()

    # plt.show() will be removed so that multiple vectors can exist on the same graph.
```

Wrapping the repeated code in a function will reduce the verbosity of the code greatly. See below for the usage.

```python
vector = np.array([3, 4])
origin = np.array([1, 1])
plot_vector(vector, origin, label='v')
plot_vector(np.array([1,3]), np.array([2,3]), label='y')
```
